## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
print(mouse_metadata.head())
print(study_results.head())

# Combine the data into a single dataset
StdResFull_df = pd.merge(mouse_metadata,study_results, on=["Mouse ID"], how = "outer")

# Display the data table for preview
StdResFull_df.head(10)
```

```python
# Checking the number of mice.
print(f"The number of unique Mouse ID in the combined data is {len(StdResFull_df['Mouse ID'].unique())}; therefore, that is the number of mice.")
print(f"The number of unique Mouse ID in the original mouse_metadata file is {len(mouse_metadata['Mouse ID'].unique())}")
if len(StdResFull_df['Mouse ID'].unique()) == len(mouse_metadata['Mouse ID'].unique()):
    print("All Mouse ID have been accounted for during the combination of the data")
else:
    print("There is a variance of Mouse ID during the combination of the data, please check the data")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
DuplicateMouse = StdResFull_df[StdResFull_df.duplicated(subset = ["Mouse ID", "Timepoint"], keep = False)]

DupMiceID = DuplicateMouse["Mouse ID"].unique()
DupMiceID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
DuplicateMouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
StdResClean_df = StdResFull_df[~StdResFull_df["Mouse ID"].isin(DupMiceID)]
StdResClean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(StdResClean_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
DRMean = pd.DataFrame([StdResClean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()])
DRMean = DRMean.swapaxes("index", "columns")
DRMean = DRMean.rename(columns={'Tumor Volume (mm3)': 'Mean'})

DRMedian = pd.DataFrame([StdResClean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()])
DRMedian = DRMedian.swapaxes("index", "columns")
DRMedian = DRMedian.rename(columns={'Tumor Volume (mm3)': 'Median'})

DRVariance = pd.DataFrame([StdResClean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()])
DRVariance = DRVariance.swapaxes("index", "columns")
DRVariance = DRVariance.rename(columns={'Tumor Volume (mm3)': 'Variance'})

DRStndDev = pd.DataFrame([StdResClean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()])
DRStndDev = DRStndDev.swapaxes("index", "columns")
DRStndDev = DRStndDev.rename(columns={'Tumor Volume (mm3)': 'Std Deviation'})

DRSEM = pd.DataFrame([StdResClean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()])
DRSEM = DRSEM.swapaxes("index", "columns")
DRSEM = DRSEM.rename(columns={'Tumor Volume (mm3)': 'SEM'})


DRStats_df = pd.merge(DRMean,DRMedian,on=["Drug Regimen"], how = "outer")
DRStats_df = pd.merge(DRStats_df,DRVariance,on=["Drug Regimen"], how = "outer")
DRStats_df = pd.merge(DRStats_df,DRStndDev,on=["Drug Regimen"], how = "outer")
DRStats_df = pd.merge(DRStats_df,DRSEM,on=["Drug Regimen"], how = "outer")
DRStats_df 
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
DRStatsAgg_df = pd.DataFrame(StdResClean_df.groupby('Drug Regimen').agg({'Tumor Volume (mm3)': 
                                                              ['mean', 'median','var', 'std', 'sem']}))
DRStatsAgg_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
DRList = ["Capomulin","Ramicane","Infubinol","Ceftamin"]
DrugRegimeSelect_df = pd.DataFrame()

for Drug in DRList:
    row = StdResClean_df.loc[StdResClean_df["Drug Regimen"] == Drug, :]
    DrugRegimeSelect_df = DrugRegimeSelect_df.append(row)

# Start by getting the last (greatest) timepoint for each mouse
GreatestTP = DrugRegimeSelect_df.groupby(["Drug Regimen", "Mouse ID"]).max()["Timepoint"]

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
LastTPPerMouse_df = pd.merge(GreatestTP, StdResClean_df, on = ("Mouse ID", "Timepoint"), how = "left")

pd.set_option('display.max_rows', None)
LastTPPerMouse_df.head(10)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
## Note: List has been created in the above cell "DRList" ##

# Create empty list to fill with tumor vol data (for plotting)
DrugTumVol_List = []
Quantiles_List = []
LowerQ_List = []
UpperQ_List = []
IQR_List = []
LwrBnd_List = []
UpprBnd_List = []
Outliers_List = []
    
# Calculate the IQR and quantitatively determine if there are any potential outliers.
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    # Determine outliers using upper and lower bounds

for drug in DRList:
    ## Create a separate dataset for each drug regimen and add to list.
    locals()[f"{drug}_df"] = LastTPPerMouse_df.loc[LastTPPerMouse_df["Drug Regimen"] == f"{drug}",:]
    DrugTumVol_List.append(locals()[f"{drug}_df"])
    
    ## Calculate the quantiles for each drug regimen
    locals()[f"{drug}Qntile"] = locals()[f"{drug}_df"]['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    Quantiles_List.append(locals()[f"{drug}Qntile"])
    
    ## Define the Lower Quartile
    locals()[f"{drug}LowQrtl"] = locals()[f"{drug}Qntile"][0.25]
    LowerQ_List.append(locals()[f"{drug}LowQrtl"])
    
    ## Define the upper Quartile
    locals()[f"{drug}UpprQrtl"] = locals()[f"{drug}Qntile"][0.75]
    UpperQ_List.append(locals()[f"{drug}UpprQrtl"])
    
    ## Define IQR
    locals()[f"{drug}IQR"] = locals()[f"{drug}UpprQrtl"] - locals()[f"{drug}LowQrtl"]
    IQR_List.append(locals()[f"{drug}IQR"])
    
    ## Calculate Lower Bound
    locals()[f"{drug}LwrBnd"] = locals()[f"{drug}LowQrtl"] - (1.5*locals()[f"{drug}IQR"])
    LwrBnd_List.append(locals()[f"{drug}LwrBnd"])
    
    ## Calculate Upper Bound
    locals()[f"{drug}UpprBnd"] = locals()[f"{drug}UpprQrtl"] + (1.5*locals()[f"{drug}IQR"])
    UpprBnd_List.append(locals()[f"{drug}UpprBnd"])
    
    ## Calculate No. Outliers
    locals()[f"{drug}Outliers"] = locals()[f"{drug}_df"].loc[(locals()[f"{drug}_df"]["Tumor Volume (mm3)"] < locals()[f"{drug}LwrBnd"]) | (locals()[f"{drug}_df"]["Tumor Volume (mm3)"]>locals()[f"{drug}UpprBnd"]),:]
    
    locals()[f"{drug}OutliersCount"] = locals()[f"{drug}Outliers"]["Mouse ID"].count()
    Outliers_List.append(locals()[f"{drug}OutliersCount"])
    
DRListStats_df = pd.DataFrame({'Drug Regime':DRList,'Lower Qrtl':LowerQ_List,'Upper Qrtl':UpperQ_List,'Lower Bound':LwrBnd_List,'Upper Bound':UpprBnd_List,"IQR":IQR_List,'No. Outliers':Outliers_List})
DRListStats_df
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
boxplot_data = [Capomulin_df['Tumor Volume (mm3)'], 
                Ramicane_df['Tumor Volume (mm3)'], 
                Infubinol_df['Tumor Volume (mm3)'], 
                Ceftamin_df['Tumor Volume (mm3)']]
boxplot_data
fig1, ax1 = plt.subplots()
ax1.set_title("Final Tumour Volume of Selected Drug Regimens")
ax1.set_ylabel("Tumour Volume (mm3)")
ax1.set_xlabel("Drug Regimen")
ax1.boxplot(boxplot_data, labels = DRList, widths = 0.8, vert = True)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Identify a Mouse ID who has been treated by Capomulin Regime
StdResClean_df.loc[StdResClean_df["Drug Regimen"]=="Capomulin", :].tail()

#Based on search, I will use Mouse ID i557
i557Study = StdResClean_df.loc[StdResClean_df["Mouse ID"]=="i557", :]
print(i557Study)

plt.plot(i557Study['Timepoint'], i557Study["Tumor Volume (mm3)"], ls = '-', marker = 'x')
plt.title('Mouse ID i557 Tumor Volume vs Timepoint (Drug Regime: Capomulin)')
plt.xlabel('Timepoints')
plt.ylabel("Tumor Volume (mm3)")
plt.xticks(i557Study['Timepoint'])
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
## Filter data to only show Drug Regime Results for Capomulin
CapomulinStdData = StdResClean_df.loc[StdResClean_df["Drug Regimen"]=="Capomulin", :]

## Group data by Mouse ID and calculate the averages
CSD_PerMouse_Mean = CapomulinStdData.groupby("Mouse ID").mean()

## Create scatter plot
X_Val = CSD_PerMouse_Mean["Weight (g)"]
Y_Val = CSD_PerMouse_Mean["Tumor Volume (mm3)"]

plt.scatter(X_Val, Y_Val)
plt.ylabel('Tumour Vol (mm3)')
plt.xlabel("Weight (g)")
plt.title('Average Mouse Tumour Volume Vs Weight On Capomulin')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(X_Val,Y_Val)
regress_val = X_Val * slope + intercept
line_eq = "y = "+str(round(slope,2)) + "x +" + str(round(intercept,2))

plt.scatter(X_Val,Y_Val)
plt.plot(X_Val, regress_val,"r-")
plt.annotate(line_eq,(0,50),fontsize= 15, color = "red")
plt.ylabel('Tumour Vol (mm3)')
plt.xlabel("Weight (g)")
plt.title('Average Mouse Tumour Volume Vs Weight On Capomulin')
print(f'R-squared value: {rvalue**2}')
plt.show()
```
